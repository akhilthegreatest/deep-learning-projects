```python
import torch
from transformers import AutoTokenizer
def preprocess_function(records):
    inputs = tokenizer(records['question1'], records['question2'], truncation=True, padding=True, return_token_type_ids=True, max_length=75, return_tensors='pt')
    inputs = {k: v.squeeze(0) for k, v in inputs.items()} # Squeeze the tensor to remove the batch dimension
    inputs['label'] = torch.tensor(records['is_duplicate'])
    return inputs
```

```python
def compute_pearson_corr(pred):
    labels = torch.from_numpy(pred.label_ids).float()
    preds = torch.from_numpy(pred.predictions.argmax(-1)).float()
    pearson_corr = torch.nn.functional.mse_loss(labels, preds).item()
    return {"pearson_corr": pearson_corr}
```

```python
from transformers import AutoModelForSequenceClassification, TrainingArguments, Trainer
import torch

# Load the pre-trained model
model = AutoModelForSequenceClassification.from_pretrained('distilbert-base-uncased')

# Define the training arguments
training_args = TrainingArguments(
    output_dir='./results',  # output directory
    num_train_epochs=3,  # total number of training epochs
    per_device_train_batch_size=64,  # batch size per device during training
    per_device_eval_batch_size=128,  # batch size for evaluation
    warmup_steps=500,  # number of warmup steps for learning rate scheduler
    weight_decay=0.01,  # strength of weight decay
    logging_dir='./logs',  # directory for storing logs
    evaluation_strategy='steps', # Evaluation and Save happens every 10 steps
    save_total_limit=5, # Only last 5 models are saved. Older ones are deleted.
    learning_rate=1e-5,
    load_best_model_at_end=True,
    fp16=True, # loads the best model when training ends
    logging_steps=1000, # log every 100 steps
    save_steps=10000, # save after every 1000 steps
    #compute_metrics_callback=compute_pearson_corr,
)

# Define the Trainer
trainer = Trainer(
    model=model,  # the instantiated Transformers model to be trained
    args=training_args,  # training arguments, defined above
    train_dataset=encoded_dataset['train'],  # training dataset
    eval_dataset=encoded_dataset['valid'],

    compute_metrics=compute_pearson_corr

      # evaluation dataset
)

# Train the model
trainer.train()
```

```python
#import torch
#from transformers import AutoTokenizer, AutoModelForSequenceClassification

#tokenizer = AutoTokenizer.from_pretrained('bert-base-cased')
#trained_model = AutoModelForSequenceClassification.from_pretrained(model_path)
trained_model = model
def check_similarity(question1, question2, debug=0):
    input_ids = tokenizer.encode(question1, question2, add_special_tokens=True, max_length=75, truncation=True, padding='max_length', return_tensors='pt')
    with torch.no_grad():
        logits = trained_model(input_ids)[0]
    predicted_class_id = int(torch.argmax(logits, axis=-1)[0])
    if predicted_class_id == 1:
        if(debug):
            print("Both questions mean the same")
        return 1
    else:
        if(debug):
            print("Both the questions are different.")
        return 0
```
